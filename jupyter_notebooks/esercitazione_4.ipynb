### Esercizio 1

__Fate, in un unico grafico, il plot delle tre funzioni $f_n(x) = \sin^2(n\,x)/(x^2 +1)$  con $n$ = 1,2,3 in 
$\,0\,\le x \le \,5$. Inserite il titolo degli assi e le legende.__

```python
#Funzione
def f(x, n):
    return np.sin(n*x)**2 / (x**2 + 1)

#Dominio
dom_f = np.linspace(0, 5, 100)

#Grafico
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))

ax.plot(dom_f, f(dom_f, 1), color='r', label=r'$ \frac{sin^2(x)}{x^2 + 1} $')
ax.plot(dom_f, f(dom_f, 2), color='g', label=r'$ \frac{sin^2(2x)}{x^2 + 1} $')
ax.plot(dom_f, f(dom_f, 3), color='b', label=r'$ \frac{sin^2(3x)}{x^2 + 1} $')

plt.legend()
```

### Esercizio 3
__Determinate il numero di intersezioni con l'asse `x`, il numero di massimi e il numero di minimi della funzione:__
$$g(x)\, = \, \left( \frac{x^5 - 5 x^2  + 7 x - 8}{x^4 +  x^2 + 1} \right).$$
__Stimate le coordinate delle intersezioni con l'asse `x`.__

```python
#Funzione
def g(x):
    return (x**5 - 5*x**2 + 7*x - 8)/(x**4 + x**2 + 1)

#Valori per grafico più grande
x = np.linspace(-50, 50, 500)
y = g(x)
```

```python
#Valori per grafico più piccolo
x2 = np.linspace(-5, 5, 100)
y2 = g(x2)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))
ax.plot(x, y, label='y=g(x)') #Funzione
ax.plot(x, np.zeros(x.size), color='k', label='y=x') #Asse delle ascisse

ax2 = plt.axes([0.55, 0.17, 0.3, 0.2])
ax2.plot(x2, y2, label='y=g(x)')
ax2.plot(x2, np.zeros(x2.size), color='k', label='y=x') #Asse delle ascisse

ax.legend()
```

Al di sopra il grafico generale. Sotto di esso il grafico con zoom sul punto di massimo e poi il grafico con zoom sul punto di minimo

Si nota che c'è un unico zero, 1 punto di massimo relativo e 1 punti di minimo relativo

zero -> 0 < x < 5/2

```python
#Valori per grafico più piccolo
x2 = np.linspace(-2.5, 0, 100)
y2 = g(x2)


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))
ax.plot(x, y, label='y=g(x)') #Funzione
ax.plot(x, np.zeros(x.size), color='k', label='y=x') #Asse delle ascisse

ax2 = plt.axes([0.55, 0.17, 0.3, 0.2])
ax2.plot(x2, y2, label='y=g(x)')
ax2.plot(x2, np.zeros(x2.size), color='k', label='y=x') #Asse delle ascisse

ax.legend()
```

### Esercizio 4
__Generate un campione di 5000 punti da una distribuzione normale, con valor medio $\mu = 1.7$ e deviazione standard $\sigma = 0.4$. Calcolate il valor medio $\mu_c$ e la deviazione standard $\sigma_c$ del campione. Determinate quanti punti del campione distano piÃ¹ di $3 \sigma_c$ da $\mu_c$. Fate l'istogramma delle frequenza del campione.__

```python
camp = np.random.normal(loc=1.7, scale=0.4, size=5000)
```

```python
#Media calcolata
camp.mean()
```

```python
#Deviazione standard calcolata
camp.std()
```

```python
#Conteggio dei campioni che verificano la condizione

#Nota: Distanza euclidea --> |c - mean| > 3*devstd

np.count_nonzero(np.abs(camp - camp.mean()) > 3*camp.std())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))
ax.hist(camp, bins=50)
```
