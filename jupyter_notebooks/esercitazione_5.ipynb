### Esercizio 1
__Disegnate la funzione $\,\, y(x)\,=\,\sqrt{2}\,x^6 - 5\, x^2 + 3\, x - 3\,$ e trovate le sue intersezioni con l'asse $x$ usando il metodo di bisezione.__

````python
def f(x):
    return (2**0.5)*x**6 - 5*x**2 + 3*x - 3

z1 = sp.optimize.bisect(f, -2, 0)
z2 = sp.optimize.bisect(f, 0, 2)

print(f"Zeri --> ({z1}, {f(z1)}) e ({z2}, {f(z2)})")
````

````python
x = np.linspace(-5, 5, 1000)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))
ax.plot(x, f(x), color='blue', label=r"$ f(x) = \sqrt{2}x^6 - 5x^2 + 3x - 3 $")

ax.scatter(z1, f(z1), color='red', label='Zeri')
ax.scatter(z2, f(z2), color='red')

ax.grid()
ax.set_ylim(-11, 11)
ax.legend()
````

### Esercizio 3
__Costruite una funzione `integ(g,A,B)`, dove g Ã¨ il nome di una funzione di una sola variabile e A e B sono array di numeri reali. `integ` deve restituire l'array di integrali:__
$$ \int_{A[i]}^{B[i]} dx \, g(x)$$
__Testate la funzione con $g(x) = x^2$, A =  [0,1,2], B = [1,2,3].__<BR>
__Nota: ciascun estremo di integrazione nella chiamata a `quad` deve essere un singolo numero reale. Non puÃ² essere un array.
In questo caso Ã¨ quindi necessario avere un ciclo `for`.__

````python
def integ(g, A, B):
    integrs = []
    
    if len(A) == len(B):        
        for estr in zip(A, B):
            #estr[0] contiene l'estremo inferiore mentre estr[1] contiene l'estremo superiore
            sol, err = sp.integrate.quad(g, estr[0], estr[1])
            
            #Aggiungo alla lista solo la soluzione e non l'errore
            integrs.append(sol)
            
    return integrs 

def g(x):
    return x**2;

A = [0, 1, 2]
B = [1, 2, 3]
            
integ(g, A, B)
````

### Esercizio 4
__Determinate le coordinate di massimi e minimi della funzione:__
$$g(x)\, = \, \left( \frac{x^5 - 5 x^2  + 7 x - 8}{x^4 +  x^2 + 1} \right).$$
__Calcolate l'integrale della funzione fra la coordinata `x1` del massimo  e quella `x2` del minimo.__

````python
def g(x):
    return (x**5 - 5*x**2 + 7*x - 8) / (x**4 + x**2 + 1)

#Definizione della funzione "opposta" per calcolare il massimo con sp.optimize.fmin
def f(x):
    return -g(x)

x2 = sp.optimize.fmin(g, 0) #Minimo
x1 = sp.optimize.fmin(f, -1) #Massimo

print(f"Massimo: ({x1}, {g(x1)})\nMinimo: ({x2}, {g(x2)})")
````

````python
x = np.linspace(-10, 10, 1000)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))
ax.plot(x, g(x), color='blue', label=r'$ \frac{x^5 - 5x^2 + 7x - 8}{x^4 + x^2 + 1} $')

ax.scatter(x1, g(x1), color='red', label='Massimo')
ax.scatter(x2, g(x2), color='green', label='Minimo')

ax.set_ylim(-15, 15)
ax.grid()
ax.legend()
````

````python
sol, err = sp.integrate.quad(g, x1, x2)

print(f"Integrale tra {x1[0]} e {x2[0]} = {sol} con un errore di {err}")
````
